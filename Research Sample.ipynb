```python
states_matrix = filtered_data_upd[['STATE_AK','STATE_AL','STATE_AZ','STATE_CO','STATE_GA','STATE_HI','STATE_IN',
                                'STATE_KY','STATE_LA','STATE_MS','STATE_MT','STATE_NC','STATE_ND','STATE_NM','STATE_NV',
                                'STATE_SC','STATE_SD','STATE_TN','STATE_WI']]

years_matrix = filtered_data_upd[[2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]]
```

```python
import numpy as np
interaction_terms = pd.DataFrame()

for state_col in states_matrix.columns:
    for year_col in years_matrix.columns:
        interaction_col_name = f"{state_col}_x_{year_col}"
        interaction_terms[interaction_col_name] = states_matrix[state_col] * years_matrix[year_col]


filtered_data_with_interactions = filtered_data_upd.join(interaction_terms)
```

```python
filtered_data_with_interactions['treat_NC'] = (filtered_data_with_interactions['STATE']=="NC").astype(float)
filtered_data_with_interactions['treat_SC'] = (filtered_data_with_interactions['STATE']=="SC").astype(float)
filtered_data_with_interactions['post_2015'] = (filtered_data_with_interactions['Year']>2015).astype(float)
```

# Regression

```python
filtered_data_with_interactions['UR'] = imputer.fit_transform(filtered_data_with_interactions[['State_UR']])
```

```python
def impute_state_trend(group):
    # Initialize the SimpleImputer for mean imputation
    imputer = SimpleImputer(strategy='mean')
    # Apply imputation only on the 'State_UR' column
    group['State_UR'] = imputer.fit_transform(group[['State_UR']])
    return group

# Group by 'make_code' and apply the imputation function
filtered_with_interactions_1 = filtered_data_with_interactions.groupby('make_code').apply(impute_state_trend)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()

filtered_with_interactions_NC['log_KWH_res'] = np.log(filtered_with_interactions_NC['KWH_res'])
filtered_with_interactions_NC['log_GSP'] = np.log(filtered_with_interactions_NC['GSP_per_capita'])
filtered_with_interactions_NC['log_CPKWH_res'] = np.log(filtered_with_interactions_NC[['CPKWH_res']])
filtered_with_interactions_NC['log_Cust_res'] = np.log(filtered_with_interactions_NC[['Cust_Res']])

NC_res = filtered_with_interactions_NC[[
    'log_Cust_res',
    'log_KWH_res', 'log_GSP','treat_NC','post_2015', 'post_x_NC']]

Price = filtered_with_interactions_NC['log_CPKWH_res'] 
```

```python
lin_reg.fit(NC_res, Price)

y_pred = lin_reg.predict(NC_res)
```

```python
coefs = lin_reg.coef_

residuals = y - y_pred
```
